Het is algemeen bekend dat de leefomstandigheden voor de mens de afgelopen honderden jaren flink verbeterd zijn. Als gevolg hiervan, is de gemiddelde levensverwachting in de afgelopen honderden jaren ook flink gestegen. Nou zijn er veel cijfers bekend over de levensverwachting van een gemiddeld persoon, maar dit waren vaak ook personen uit de lagere sociale klassen. Natuurlijk is niet elke kunstschilder bekend (en daardoor welvarend) geweest in de periode dat hij of zij leefde, maar gemiddeld genomen zou er best wel eens een verschil kunnen zijn tussen de levensverwachting van een gemiddelde burger en de levensverwachting van een gemiddelde kunstschilder gedurende de afgelopen honderden jaren. Of dit het geval is, wordt beantwoord door onderstaande grafiek. 

```python
# Thomas 2
# SPARQL query draaien
df = wikidata.query('''
# List of all Dutch famous painters, their date of birth, place of birth, and corresponding coordinates (for map view)
SELECT DISTINCT ?schilderLabel ((YEAR(xsd:dateTime(?geboortedag))) AS ?geboortejaar) ?geboorteplaatsLabel (YEAR(?sterfdag) - year(?geboortedag) AS ?sterfleeftijd)
WHERE {
    ?schilder wdt:P101 wd:Q11629 .
    ?schilder wdt:P569 ?geboortedag .
    ?schilder wdt:P570 ?sterfdag .
    ?schilder wdt:P569 ?geboortedag .
    ?schilder wdt:P19 ?geboorteplaats .
    ?geboorteplaats wdt:P17 wd:Q55 .
    
SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl" }
}
''')

# filteren op sterfleeftijd boven de 100 (deze kloppen niet)
df['geboortejaar'] = pd.to_numeric(df['geboortejaar'])
df['sterfleeftijd'] = pd.to_numeric(df['sterfleeftijd'])
df = df[df.sterfleeftijd < 100]

# groeperen op jaartal om gemiddelde sterfleeftijd te krijgen
df = df.groupby('geboortejaar').mean()
df.reset_index(level=0, inplace=True)

# plotten
fig1 = px.scatter(df, x="geboortejaar", y="sterfleeftijd", trendline="ols", title='Gemiddelde sterfleeftijd Nederlandse schilders t.o.v. de tijd')
fig1.show()
```

Natuurlijk hoeft het zijn van een kunstschilder geen stempel te zijn op deze personen. Kunstschilders kunnen hiernaast heel goed nog andere functies hebben, en dat wordt ook duidelijk aan de hand van onderstaand plot. Er is te zien dat van alle kunstschilders, naast het feit dat natuurlijk 100% kunstschilder is, er ook een groot gedeelte andere functies hebben. Sommige kunstschilders hebben hiernaast heel vergelijkbare functies zoals tekenaar of kunstenaar, maar andere kunstschilders hebben  hiernaast hele andere functies zoals advocaat, glazenier of vertaler. 

```python
# Thomas 3
df = wikidata.query('''
# List of all Dutch famous painters, their date of birth, place of birth, and corresponding coordinates (for map view)
SELECT DISTINCT ?schilderLabel ?alleberoepenLabel
WHERE {
    ?schilder wdt:P101 wd:Q11629 .
    ?schilder wdt:P19 ?geboorteplaats .
    ?geboorteplaats wdt:P17 wd:Q55 .
    ?schilder wdt:P106 ?alleberoepen
SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],nl" }
}
''')

df = df['alleberoepenLabel'].value_counts().to_frame()
df.reset_index(level=0, inplace=True)
df = df.rename(columns={"index": "beroep"})
df['percentage'] = df['alleberoepenLabel'] / df['alleberoepenLabel'][0] * 100

fig2 = px.histogram(df, x='beroep', y='percentage', title='beroepen van Nederlandse kunstschilders in percentage van het aantal kunstschilders', histfunc='sum')


fig2.show()
```
